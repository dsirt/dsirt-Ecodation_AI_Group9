```python
mycount = 0
while mycount < 9:
    print('The count is:', mycount)
    mycount += 1
print("Bye Bye!")
```

```python
#while-else
mycounty = 0
while(mycounty < 7):
    print(mycounty, "is less than 7")
    mycounty += 1
else:
    print(mycounty, "is not less than 7")
```

```python
x = -10
y = 10
while x <= y:
    print("X is now: ", x)
    x = x + 1
    while x < 0:
        print("The value of x:", x)
        x = x + 1
```

```python
fruits = ['banana', 'apple', 'mango']
for fruit in fruits:
    print('Current fruit :', fruit)
print('Bye Bye!')
```
